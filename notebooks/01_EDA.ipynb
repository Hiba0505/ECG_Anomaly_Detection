```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("../data/raw/mitbih_train.csv", header=None)
test  = pd.read_csv("../data/raw/mitbih_test.csv",  header=None)

X = train.iloc[:, :187].values
y = train.iloc[:, 187].values.astype(int)
print("Shape train:", train.shape)
```

```python
classes = {0:"Normal", 1:"Supraventriculaire", 2:"Ventriculaire", 3:"Fusion", 4:"Autre"}
counts = train[187].value_counts().sort_index()

plt.figure(figsize=(8,4))
plt.bar([classes[i] for i in counts.index], counts.values, color="steelblue")
plt.title("Distribution des classes MIT-BIH")
plt.ylabel("Nombre de battements")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(5, 1, figsize=(14, 12))
for cls in range(5):
    idx = np.where(y == cls)[0][0]
    axes[cls].plot(X[idx], color="steelblue", linewidth=1.2)
    axes[cls].set_title(f"Classe {cls} — {classes[cls]}")
    axes[cls].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

# Sous-échantillon pour rapidité
idx = np.random.choice(len(X), 3000, replace=False)
X_sample, y_sample = X[idx], y[idx]

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_sample)

plt.figure(figsize=(8,6))
for cls in range(5):
    mask = y_sample == cls
    plt.scatter(X_pca[mask,0], X_pca[mask,1], label=classes[cls], alpha=0.5, s=10)
plt.title("PCA — Séparabilité des classes ECG")
plt.legend()
plt.show()
```
